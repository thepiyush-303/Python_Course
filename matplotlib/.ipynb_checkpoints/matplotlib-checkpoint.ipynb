```python
labels=['A','B','C']
values=[4,7,8]
bargraph=plt.bar(labels,values)
bargraph[0].set_hatch('/')
patterns=['/','o','-']
for bar in bargraph:
    bar.set_hatch(patterns.pop(0))
plt.figure(figsize=(5,3),dpi=10)
plt.show()
```
