```python
labels=['A','B','C']
values=[4,7,8]
bargraph=plt.bar(labels,values)
bargraph[0].set_hatch('/')
patterns=['/','o','-']
for bar in bargraph:
    bar.set_hatch(patterns.pop(0))
plt.figure(figsize=(1,3))
plt.show()
```

```python
 gas=pd.read_csv("gasprices.csv")
```

```python
gas
```

```python
plt.title("Gas prices of different countries in USD")
# plt.plot(gas.Year,gas.USA,label="USA",'b--')
# plt.plot(gas.Year,gas.Canada,label="Canada", 'g--')
# plt.plot(gas.Year,gas.Australia,label="Australia", 'y**')
for country in gas:
    print(country)
    if country != 'Year':
        plt.plot(gas.Year,gas[country],label=country)
    
plt.legend()
plt.xticks(gas.Year[::3].tolist()+[2011])
plt.show()
plt.savefig("gas_price_fig.png",dpi=300)
```
